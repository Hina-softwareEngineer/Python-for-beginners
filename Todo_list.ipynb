```python
array=[]


def show():
    i=1
    for elements in array:
        print(i,". ",elements)
        i=i+1

def add():
    inp=input("Enter what do you want to add?")
    array.append(inp)
    print(inp, " has been added.")

def delete():
    inp=input("Which name do you want to delete?")
    array.remove(inp)
    print(inp," has been removed.")

def update():
    inp=int(input("Which element do you want to update. Enter position."))
    element=input("Which is the new element to be replaced at your given postition.")
    array[inp-1]=element
        
def deleteall():
    array.clear()


while True:
    show()
    
    print("\n\n=============================")
    print("Which option do you want to choose?")
    print("1. Display a list")
    print("2. Add an element")
    print("3. Update an element")
    print("4. Delete an element")
    print("5. Delete all element")
    print("6. Exit")

    opt=input("Enter your option:")
    if opt=="1":
        show()
    elif opt=="2":
        add()
    elif opt=="3":
        update()
    elif opt=="4":
        delete()
    elif opt=="5":
        deleteall()
    elif opt=="6":
        break
    else:
        print("INvalid INput")

```
